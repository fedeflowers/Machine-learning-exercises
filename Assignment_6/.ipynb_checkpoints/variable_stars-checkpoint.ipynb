# With right files

```python
import pandas as pd
# reading csv files
train = pd.read_csv('VSTrain_formatted.txt', sep=',', header=None)
test =  pd.read_csv('VSTest_formatted.txt',sep=',', header=None)
```

```python
train
```

```python
import numpy as np
a, b = np.unique(train[61], return_counts=True)
a,b
frequencies = b/len(train)
frequencies
```

```python
from matplotlib import pyplot as plt
plt.xlabel("classes")
plt.ylabel("frequencies")
plt.plot(frequencies)
```

```python
freqs = {}
for i,f in enumerate(frequencies):
    freqs[i] = f
    
freqs
```

```python
classes_to_delete = np.where(b<65)
classes_to_delete
```

```python
len(test)
```

```python
indixes_to_drop = []
for i in range(len(train)):
    if int(train.iloc[i,:][61]) in np.array(classes_to_delete):
        indixes_to_drop.append(train.index[i])
```

```python
0 in np.array(classes_to_delete)
```

```python
train_trimmed = train.drop(indixes_to_drop)
```

```python
train_trimmed
```

```python
a, b = np.unique(train_trimmed[61], return_counts=True)
```

```python
a,b
```

```python
train_trimmed.mean(axis=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
classes = train_trimmed.pop(61)
standardized_train = scaler.fit_transform(train_trimmed)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
transformed_data = pca.fit_transform(standardized_train)
```

```python
plotDF = pd.DataFrame(data = transformed_data, columns = ['pc1', 'pc2'])
```

```python
classes = dict(classes)
```

```python
final_classes = []
for k in classes:
    final_classes.append(classes[k])
```

```python
finalDf = pd.concat([plotDF, pd.Series(final_classes)], axis = 1)
```

```python
finalDf.rename(columns = {0:'classes'}, inplace = True)
```

```python
finalDf
```

```python
import seaborn as sns

sns.relplot(data=finalDf, x='pc1', y='pc2', hue='classes', palette='tab10', kind='scatter')
```

# Clustering

```python
from sklearn.cluster import KMeans as km

kmeans = km(n_clusters=4, init='random', n_init = 1)
kmeans_plus_plus = km(n_clusters=4, init='k-means++', n_init= 1)
```

```python
kmeans.fit(plotDF)
```

```python
kmeans_plus_plus.fit(plotDF)
```

```python
pd.concat([pd.DataFrame(kmeans.cluster_centers_), pd.Series(data = [0,0,0,0])],axis=1)
```

```python
kmeans_plus_plus.cluster_centers_
```

```python
sns.scatterplot(data=finalDf, x='pc1', y='pc2', hue='classes', palette='tab10')
```

# Work on dataframes

```python
kmeans_centers_pd = pd.DataFrame(kmeans.cluster_centers_)
kmeans_centers_pd.rename(columns = {0:'x', 1:'y'}, inplace = True)
kmeans_centers = pd.concat([kmeans_centers_pd, pd.Series(data = ['center','center','center','center'])],axis=1)
kmeans_centers.rename(columns = {0:'hue'}, inplace = True)
kmeans_centers
```

```python
import seaborn as sns
#KMEANS
ax = sns.scatterplot(data=finalDf, x='pc1', y='pc2', hue='classes', palette='tab10')
sns.scatterplot(kmeans_centers, x='x', y= 'y',s=200, hue='hue',palette='Set2')
```

# Kmeans++

```python
kmeans_centers_pd = pd.DataFrame(kmeans_plus_plus.cluster_centers_)
kmeans_centers_pd.rename(columns = {0:'x', 1:'y'}, inplace = True)
kmeans_centers = pd.concat([kmeans_centers_pd, pd.Series(data = ['center','center','center','center'])],axis=1)
kmeans_centers.rename(columns = {0:'hue'}, inplace = True)
kmeans_centers
```

```python
import seaborn as sns
#KMEANS ++
ax = sns.scatterplot(data=finalDf, x='pc1', y='pc2', hue='classes', palette='tab10')
sns.scatterplot(kmeans_centers, x='x', y= 'y',s=200, hue='hue',palette='Set2')
```
