```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets

import matplotlib.pyplot as plt
%matplotlib inline
```

Load handwritten digits data:

```python
digits = datasets.load_digits()
imshape = digits.images[0].shape  # The 2D image shape
X = digits.images.reshape(digits.images.shape[0],-1)  # Image
```

Do the PCA:

```python
pca = PCA()
pca.fit(X)
eigenvalues = pca.singular_values_**2
explained_var_comp = pca.explained_variance_ratio_
```

```python
plt.yscale('log')
plt.plot(eigenvalues / np.sum(eigenvalues))
```

Find out if 10 components are enough to explain 80% of the variance:

```python
np.sum(explained_var_comp[:10]) #not enough
```

```python
plt.plot(np.cumsum(explained_var_comp[:10]))
```

```python
plt.plot(np.cumsum(explained_var_comp)) #explained variance, unnecessary
```

Plot the eigenspectrum:

```python
plt.plot(explained_var_comp)#eigenspectrum #unnecessary
```
